```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
file_path='/content/IRIS[1].csv'
df = pd.read_csv(file_path)
print(df.head())
```

```python
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
plt.figure(figsize=(12, 8))
for i, feature in enumerate(features):
    plt.subplot(2, 2, i+1)
    sns.scatterplot(x=df[feature], y=[0]*len(df), hue=df['species'], palette='Set2')
    plt.title(f'Univariate Analysis of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Constant Value (0)')
plt.tight_layout()
plt.show()

# Step 5: Bivariate Analysis
bivariate_pairs = [
    ('sepal_width', 'petal_width'),
    ('sepal_length', 'petal_length')
]
```

```python
plt.figure(figsize=(10, 5))
for i, (x_feature, y_feature) in enumerate(bivariate_pairs):
    plt.subplot(1, 2, i+1)
    sns.scatterplot(data=df, x=x_feature, y=y_feature, hue='species', palette='Set1')
    plt.title(f'{x_feature} vs {y_feature}')
plt.tight_layout()
plt.show()
```

```python
# b. Create pairplot with KDE on diagonal
sns.pairplot(df, hue='species', diag_kind='kde', palette='husl')
plt.suptitle('Multivariate Analysis: Pairplot with KDE', y=1.02)
plt.show()
```
